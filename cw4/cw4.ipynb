# Zad 1.
###### Stworzyć funkcję własną, która będzie zwracać listę argumentów oraz odpowiadających im wartości funkcji y=ax^2+bx+c

```python
%matplotlib inline
import matplotlib.pylab as pylab
import math
import numpy as np
```

```python
def listy(a,b,c):
    x = [0.2*i for i in range(-100,100)]
    y = [a*el**2+b*el+c for el in x]  
    
    return x,y


x,y = listy(8,3,8)
#pylab.figure(figsize=(10,5))
pylab.grid(True)
pylab.plot(x,y)
```

# Zad 2.
###### Przerób funkcję z zadania poprzedniego do postaci kanonicznej: y =a(x-p)^2+q, p = -b/2a, q = - Delta/4a

```python
def kanoniczna(a,p,q):
    x = [i for i in range(-10,10)]
    y = [a*(el-p)**2+q for el in x]
    return x,y
```

```python
%matplotlib notebook
x,y = kanoniczna(5,-4,-150)
pylab.grid(True)
pylab.plot(x,y)
```

# Zad 3.

```python
%matplotlib inline
for i in range(1,11):
    a= i**3.5
    p= a/3
    q= p**0.5
    x,y = kanoniczna(a,p,q)
    pylab.plot(x,y)
    pylab.savefig("wykres{}".format(i))
    pylab.show()
```
