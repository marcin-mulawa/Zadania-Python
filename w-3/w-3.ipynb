###### Utwórz funkcję, która będzie generować listy danych do wykreślenia w oparciu o:
a) fukcję liniową ax+b <br>
b) funkcję kwadratową ax^2+bx+c <br>
c) funkcję odwrotnie-potęgową a/x^n <br>

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def f_line(a,b):
    x = [i for i in range(-100,100)]
    y = [a*x+b for x in x] 
    return x,y
```

```python
def f_kwadratowa(a,b,c):
    x = [i for i in range(-100,100)]
    y = [a*(x**2)+b*x+c for x in x] 
    return x,y
```

```python
def f_odwrotnie(a,n):
    x = [i for i in range(-100,100)]
    y = [a/x**n if x!=0  else None for x in x]
    return x,y
```

```python
x,y = f_line(-1,5)
plt.plot(x,y)
```

```python
x,y = f_kwadratowa(2,3,10)
plt.grid(True)
plt.plot(x,y)
```

```python
%matplotlib inline
x,y = f_odwrotnie(22,7)
plt.grid(False)
plt.plot(x,y)
plt.show()
```
